# MX1

## read

```typescript
const xmlText = await Deno.readTextFile("M-1-0-1.1.MX1");
```

## parse

```typescript
const parser = new XMLParser({
    ignoreAttributes: false,
    attributeNamePrefix: "",//@@_",
    parseAttributeValue: true,
    trimValues: true
  }
  );
const jsonObj = parser.parse(xmlText);
```

## mx1Df

```typescript
let items = jsonObj.DocumentElement.XL1;
  // Ensure it's an array
if (!Array.isArray(items)) {
    items = [items];
}
let mx1Df = pl.DataFrame(items) 
```

```typescript
// console.log(mx1Df.toString());
console.table(mx1Df.toRecords());


const sortedUnique = mx1Df.getColumn("DATATYPE").unique().sort().toArray();
console.log(sortedUnique);
```

#### annotations

```typescript
mx1Df = mx1Df.withColumns(
    pl.col('DATALENGTH').div(pl.col('DATATYPELENGTH'))
    .cast(pl.Int64)
    .alias('NOfItems')
);
```

```typescript
mx1Df = mx1Df.withColumns(
  pl.when(
    pl.col('NOfItems')
    .gt(1).or(
      pl.col('OBJTYPE_PK').str.lengths().lt(3)
      .and(pl.col('OBJTYPE').cast(pl.Utf8).str.contains("^ALLG"))
    )
  )
  .then(pl.lit(true))
  .otherwise(pl.lit(false))
  .alias('isVectorChannel')
);
```

```typescript
mx1Df = mx1Df.withColumns(
    pl.when(
      pl.col('isVectorChannel')
        .and(pl.col('FLAGS').gtEq(4))
        .and(pl.col('FLAGS').cast(pl.Int64)
    .div(4)
    .floor()
    .modulo(2)
    .eq(1))
    )
    .then(pl.lit(true))
    .otherwise(pl.lit(false))
    .alias('isVectorChannelMx2')
);
```

```typescript
mx1Df = mx1Df.withColumns(
    pl.when(
      pl.col('isVectorChannelMx2')
        .and(pl.col('DATATYPE').cast(pl.Utf8).str.contains("^RVEC"))
    )
    .then(pl.lit(true))
    .otherwise(pl.lit(false))
    .alias('isVectorChannelMx2Rvec')
);  
```

```typescript
console.table(mx1Df.toRecords());
```

# MX2

## read and parse to mx2Df

```typescript
const mx2File = await Deno.open("M-1-0-1.MX2");

const mx2Ar = [];
  
try {
    while (true) {

      // Header
      const bufferHeader = new Uint8Array(64); 
      const bytesReadHeader = await mx2File.read(bufferHeader);
      if (bytesReadHeader === null) break; // EOF
      if (bytesReadHeader !== 64) {
        console.warn("Header: bytesReadHeader !== 64?! - break...");
        break; 
      }      
      const recordHeaderUnpacked = bufferHeader.slice(0,bytesReadHeader);
      const recordHeader = unpackRecord(recordHeaderUnpacked);
      recordHeader.NOfObjs = Math.floor(recordHeader.DataLength / recordHeader.DataTypeLength);

      //console.log(recordHeader);

      // Data
      const bufferContent = new Uint8Array(recordHeader.DataLength); 
      const bytesReadContent = await mx2File.read(bufferContent);
      const recordContentUnpacked=bufferContent.slice(0,bytesReadContent);

      if (recordHeader.DataType === 'CHAR') {
        const decoder = new TextDecoder("utf-8");
        const items=[];
        for (let i = 0; i < recordHeader.NOfObjs; i++) {        
          const start=i*recordHeader.DataTypeLength;
          const end=start+recordHeader.DataTypeLength;
          const item = decoder.decode(recordContentUnpacked.subarray(start, end)).replace(/\0.*$/, "").replace(/\s+$/, "");
          items.push(item);
          //console.log(`CHAR Item ${i}: ${item}`);
          //break;
        }
        recordHeader.Data=items
      }
      else if (recordHeader.DataType === 'INT4') {
        const view = new DataView(recordContentUnpacked.buffer, recordContentUnpacked.byteOffset, recordContentUnpacked.byteLength);
        const items=[];
        for (let i = 0; i < recordHeader.NOfObjs; i++) {        
          const start=i*recordHeader.DataTypeLength;
          const item = view.getInt32(start, true); // true = little-endian
          items.push(item);
          //console.log(`INT4 Item ${i}: ${item}`);
          //break;
        }    
        recordHeader.Data=items    
      }
      else {
        console.warn(`DataType ${recordHeader.DataType} not yet supported - break...`);
        break;
      }
      mx2Ar.push(recordHeader);

    }
  } finally {
    mx2File.close();
  }

  const mx2Df = pl.DataFrame(mx2Ar);
  console.log(mx2Df.toString());

  console.log("Schema:", mx1Df.schema);
  console.log("Shape:", mx1Df.shape);
  console.log("Columns:", mx1Df.columns);
  console.log("Dtypes:", mx1Df.dtypes);

```

# MXS

## read

```typescript
const lastMx1 = mx1Df.tail(1).toRecords()[0];
const mxRecordLength = lastMx1.DATAOFFSET+lastMx1.DATALENGTH;
const buffer = new Uint8Array(mxRecordLength); 



let records=[];




const mxsFile = await Deno.open("M-1-0-1.1.MXS");


  
try {
    //
    while(true) {
        const bytesRead = await mxsFile.read(buffer);
        if (bytesRead === null) break; // EOF
        if (bytesRead !== mxRecordLength) {
            console.warn(`bytesRead (${bytesRead}) !== mxRecordLength (${mxRecordLength})?! - break...`);
            break;
        }
        console.log(`Read ${bytesRead} bytes from MXS file.`);

        const recordUnpacked = buffer.slice(0,bytesRead);
        const view = new DataView(recordUnpacked.buffer, recordUnpacked.byteOffset, recordUnpacked.byteLength);
        const decoder = new TextDecoder("utf-8");

        let record={};

        for (const row of mx1Df.toRecords()) {

            if (row.isVectorChannel) {
                //console.log("isVectorChannel - skip for now...");
                continue;
            }
        
            const key=`${row.OBJTYPE}~${row.ATTRTYPE}~${row.NAME1}~${row.NAME2}~${row.NAME3}~${row.OBJTYPE_PK}`;
            if (row.DATATYPE === 'CHAR') {
                    const item = decoder.decode(recordUnpacked.subarray(row.DATAOFFSET, row.DATAOFFSET + row.DATALENGTH)).replace(/\0.*$/, "").replace(/\s+$/, "");
                    record[key]=item;
            }
            else if (row.DATATYPE === 'INT4') {
                    const item = view.getInt32(row.DATAOFFSET, true); // true = little-endian
                    record[key]=item;
            }
            else if (row.DATATYPE === 'REAL') {
                    const item = view.getFloat32(row.DATAOFFSET, true); // true = little-endian
                    record[key]=item;
            }
            
        }
        records.push(record);


            
    }
} finally {
    mxsFile.close();
}



```

```typescript
let mxsDf = pl.DataFrame(records);
//#console.log(mxsDf.toString());
```

```typescript
  /*
  console.log("Schema:", mxsDf.schema);
  console.log("Shape:", mxsDf.shape);
  console.log("Columns:", mxsDf.columns);
  console.log("Dtypes:", mxsDf.dtypes);
  */
  //console.table(mxsDf.toRecords());
```

```typescript
mxsDf = mxsDf.withColumns(
  pl.col("ALLG~TIMESTAMP~~~~5108484491127470798")
    .str.strptime(pl.Datetime, "%Y-%m-%d %H:%M:%S%.6f%:z", {
      strict: false
    })
    .alias("Timestamp")
);
```
